```python
import pandas as pd
```

```python
# let us read the json file
general_training = pd.read_json('response_train.json')
```

```python
general_training.head()
```

```python
# let us drop passages, query_id and query_type and wellFormedAnswers
general_training.drop(['question', 'response', 'acceptable?', 'category','question_category'], axis=1, inplace=True)
```

```python
general_training
```

```python
# let us write a .txt file with question_text and then space line and then answer
# let us open a file
f = open('general_training.txt', 'w')

# let us write the question_text and then the answer
for i in range(len(general_training)):
    f.write(general_training.iloc[i]['question_en'] + '\n\n')

# let us close the file
f.close()
```
